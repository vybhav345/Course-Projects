# Monte Carlo Simulations of 1D & 2D Random Walks

This project explores Monte Carlo simulations of random walks, including 1D unbiased and biased walks, 2D random walks, and mean-square displacement (MSD) analysis. The study investigates how random motion evolves over time and validates theoretical predictions of diffusive behavior. Additionally, a linear congruential generator (LCG) is implemented to generate pseudo-random numbers, analyzing its periodicity. The results are visualized through matplotlib plots showcasing particle trajectories and statistical trends.

```python
import numpy as np
import matplotlib.pyplot as plt
import random

def Randomwalk1D(n): #n is number of jumps
  #initialization
   step_no = 0
   x = 0

  #append variable initialization
   total_steps = [0]
   X = [0]
   X2 = [0]

   for i in range (1,n+1):
       step = np.random.uniform(0,1) #produces random number between 0 and 1

       if step < 0.5:
           step_no=step_no+1
           x=x-1

       if step > 0.5:
           step_no=step_no+1
           x=x+1

       total_steps.append(step_no)
       X.append(x)
       X2.append(x**2)

   return [total_steps,X,X2]

#1D random walk for 1000 jumps
Randwalk = Randomwalk1D(1000)

#plot1
plt.plot(Randwalk[0],Randwalk[2],'b', label='x2')
plt.title("1D Random Walks with 1000 jumps")
plt.xlabel("number of stpes")
plt.legend()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Define parameters
n_simulations = 1000
n = 1000 #n is number of jumps

# Initialize variables to store results
x2_means = np.zeros(n+1)

# Run multiple simulations
for i in range(n_simulations):
    position = 0
    x2_values = np.zeros(n+1)
    for j in range(n+1):
        x2_values[j] = position**2
        if j < n:
            position += random.choice([-1, 1])
    x2_means = x2_means + x2_values

# Average the results
x2_means = x2_means/n_simulations

# Plot the mean-square displacement
steps = list(range(0,1001))
plt.plot(steps, x2_means)
plt.plot(steps, steps, linestyle='dashed', color='red')
plt.xlabel('Step number')
plt.ylabel('MSD')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

#initialize and define parameters
n = 1000
x = np.zeros(n)
y = np.zeros(n)

direction=["up","down","left","right"]

for i in range(1, n):
    step = random.choice(direction)
    if step == "right":
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1]
    elif step == "left":
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1]
    elif step == "up":
        x[i] = x[i - 1]
        y[i] = y[i - 1] + 1
    else:
        x[i] = x[i - 1]
        y[i] = y[i - 1] - 1
plt.title("2D Random Walks with 1000 jumps")
plt.plot(x, y)
plt.show()
```

```python
# Set the initial seed, multiplier, and modulus
seed = 71
a = 18
m = 167
n = 200

# Generate the series of random numbers
x = [seed]
for i in range(1, n):
    x.append((a * x[i-1]) % m)

# Print the generated series
print(x)

# Find the period of the generated series
def findperiod(f, minlength=1, repetitions=2):
  for i in (range(minlength, n)):
        for rep in range(1, repetitions):
            for j in range(i):
                if f[j] != f[i*rep+j]: break
            else: return i

# Print the period
print("Period:",findperiod(x) )

```

```python
# Try different values of a and find the maximum period
max_period = 0
max_a = None
for a in range(2,m):
    x = [seed]
    period = 1
    while True:
        next_num = (a * x[period-1]) % m
        if next_num == seed:
            if period > max_period:
                max_period = period
                max_a = a
            break
        x.append(next_num)
        period += 1
        if period > 1000:
            break

# Print the maximum period and the corresponding value of 'a'
print("Maximum period:", max_period)
print("Value of 'a':", max_a)
```

```python
import numpy as np

# Set the initial positions of the sailors
sailor1 = np.array([0, 0])
sailor2 = np.array([5, 5])

def move(sailor):
    """Move the sailor to a neighboring cell in the grid."""
    dx, dy = np.random.choice([-1, 0, 1], size=2)
    sailor = sailor + np.array([dx, dy])
    sailor = sailor % 10  # periodic boundary conditions

def simulate():
    """Simulate the random walks of two sailors until they bump into each other."""
    steps = 0
    while not np.array_equal(sailor1, sailor2):
        move(sailor1)
        move(sailor2)
        steps = steps + 1
    return steps

# Run multiple simulations and compute the average number of steps
num_simulations = 1000
total_steps = 0
for i in range(num_simulations):
    total_steps = total_steps + simulate()
average_steps = total_steps / num_simulations

print("Average number of steps:", round(average_steps))
```
